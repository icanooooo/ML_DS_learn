```python
from matplotlib import pyplot as plt
```

# Simple Plot Chart

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.3, 543.3, 1075.9, 2862.5, 5976.6, 10289.7, 14958.3]

plt.plot(years, gdp, color='green', marker='o', linestyle='solid')
plt.title('Nominal GDP')

plt.ylabel('billions of $')
plt.show()
```

# Bar Chart

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
oscars = [5, 11, 3, 8, 10]

plt.bar(movies, oscars)
plt.ylabel("No. Of Oscars")
plt.xlabel("Movies")

plt.show()
```

```python
from collections import Counter

grades = [83,95,91, 87, 70, 83, 93, 100, 30, 55, 54, 0]
# Function untuk mengubah ke bilangan bawah terdekat
decile = lambda grade: (grade // 10) * 10 

histogram = Counter(decile(grade) for grade in grades) # Fungsi counter intuk meng'count' suatu unique values di suatu list

plt.bar([x for x in histogram.keys()],
        histogram.values(), 8)

plt.axis([-5, 105, 0, 5])
plt.xlabel("Decile")
plt.ylabel("# of Students")
plt.title("Distribution of Exam 1 Grades")
plt.show()
```

```python
# Membanding perubahan axis

mentions = [500, 505]
year = [2013, 2014]

plt.bar(year, mentions, 0.8)
plt.ylabel("count")
plt.xticks(year) # untuk custom ticks dari x axis
plt.ticklabel_format(useOffset=False)
plt.axis([2012.5, 2014.5, 499, 506]) # x dari 2012.5-2014.5, y dari 499-506
plt.show()
```

```python
plt.bar(year, mentions, 0.8)
plt.ylabel("count")
plt.xticks(year)
plt.ticklabel_format(useOffset=False)
plt.axis([2012.5, 2014.5, 0, 600]) # Jika dijauhkan bisa terlihat lebih tidak signifikan
plt.show()
```

# Line Plot

```python
variance = [2**x for x in range(9)]
bias_squared = list(reversed(variance))
total_error = [x+y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

plt.plot(xs, variance, 'g-', label='variance')
plt.plot(xs, bias_squared, 'r-.', label='bias^2')
plt.plot(xs, total_error, 'b:', label='total error')

plt.legend(loc=9) # top center for legend
plt.xlabel("model complexity")
plt.title("The Bias-Variance Tradeoff")

plt.show()
```

# Scatterplots

```python
friends = [70, 67, 63, 60, 62, 54, 52]
minutes = [175, 170, 200, 120, 165, 145, 100]
labels = [x for x in 'abcdefgh']

plt.scatter(friends, minutes)

# Menambahkan anotasi di masing-masing points
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
                 xy=(friend_count, minute_count),
                 xytext=(5, -5),
                 textcoords='offset points')
    
plt.title('daily minutes vs. no of friend')

plt.show()
```
